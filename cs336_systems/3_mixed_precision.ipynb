```python
import torch.nn as nn
class ToyModel(nn.Module):
    def __init__(self, in_features, out_features):
        super().__init__()
        self.fc1 = nn.Linear(in_features, 10, bias = False)
        self.ln  = nn.LayerNorm(10)
        self.fc2 = nn.Linear(10, out_features, bias = False)
        self.relu = nn.ReLU()
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.ln(x)
        x = self.fc2(x)
        return x
```

```python
batch_size, seq_len, vocab_size, d_k = 500, 100, 50, 1000
x = torch.randn(batch_size, seq_len, d_k)
model = ToyModel(in_features = d_k, out_features=20)
import cProfile, io, pstats
from pstats import SortKey



profiler = cProfile.Profile()
profiler.enable()
with torch.autocast(device_type = 'cpu', dtype = torch.float16):
    y = model(x)

profiler.disable()
# Create a string buffer to capture stats
s = io.StringIO()
ps = pstats.Stats(profiler, stream=s)
ps.sort_stats(SortKey.TIME)  # Sort by cumulative time

# Print statistics
print("="*80)
print("cProfile Results - Top 20 functions by cumulative time")
print("="*80)
ps.print_stats(20)
print(s.getvalue())
```
