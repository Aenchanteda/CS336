```python
#2. __repr__()方法与print方法的区别？
'''
repr(obj) 触发 __repr__()，返回面向调试的、尽量精准且可重建对象的字符串；
print(obj) 调用 str(obj)（若未实现则退回 __repr__()），面向用户展示，可读性更高，不一定可重建对象
'''
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __repr__(self):
        return f"Point(x={self.x}, y={self.y})"

p = Point(2, 3)
clone = eval(repr(p))  # 通过 repr 返回的字符串重建对象（eval将字符串当作代码执行，相当于再次调用p，从而得到一个拥有相同x、y的新对象）
```

```python
#4.encode和decode函数：UTF-8和Unicode string转换
test_string = "hello, 你好好"
utf_8_encoded = test_string.encode("utf-8")#ASCII段直接显示，中文非ASCII段以十六进制转义
print(type(utf_8_encoded))
list(utf_8_encoded)

print(len(test_string))
print(len(utf_8_encoded))
print(type(utf_8_encoded.decode("utf-8")))

# .encode()使用主体是str,结果是bytes类型，.decode()使用主体是bytes类型，结果是str
```
